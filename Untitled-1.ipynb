```python
from selenium import webdriver
#import time
from selenium.webdriver.common.by import By

#Abrir o navegador
navegador = webdriver.Chrome()

navegador.get('https://pokemondb.net/type')

#colocar navegador em tela cheia
navegador.maximize_window()

# Encontra todos os td que possuem atributo title
elementos = navegador.find_elements(By.CSS_SELECTOR, "td[title]")

# Cria uma lista com os valores do atributo title
titulos = [elemento.get_attribute("title") for elemento in elementos]

navegador.quit()
```

```python
print(len(titulos))
print(titulos)
```

```python
import pandas as pd

mapa_efetividade = {
    "normal effectiveness": 1,
    "not very effective": 0.5,
    "no effect": 0,
    "super-effective": 2
}

dados = []

for titulo in titulos:
    atacante, resto = titulo.split(" → ")
    defensor, resultado = resto.split(" = ")

    dados.append({
        "ATACANTE": atacante,
        "DEFENSOR": defensor,
        "EFETIVIDADE": resultado,
        "MULTIPLICADOR": mapa_efetividade[resultado]
    })

df = pd.DataFrame(dados)

print(df)

df.to_csv("dataframes/efetividade.csv", index=False)
```

```python
from bs4 import BeautifulSoup

# Selenium abre a pg
navegador = webdriver.Chrome()
navegador.get("https://pokemondb.net/pokedex/all")

# Selenium captura o HTML da tabela
tabela = navegador.find_element(By.TAG_NAME, "table")
html_tabela = tabela.get_attribute("outerHTML")

# BeautifulSoup processa o HTML
soup = BeautifulSoup(html_tabela, "html.parser")

dados = []

for tr in soup.find_all("tr"):
    linha = [td.get_text(strip=True) for td in tr.find_all("td")]

    if linha:
        dados.append(linha)

navegador.quit()
print(dados)
```

```python
import re

colunas = [
    'NUMERACAO',
    'NOME',
    'TIPO',
    'TOTALSTATS',
    'HP',
    'ATAQUE',
    'DEFESA',
    'ATAQUE_ESPECIAL',
    'DEFESA_ESPECIAL',
    'VELOCIDADE'
]

df = pd.DataFrame(dados, columns=colunas)

df['TIPO'] = df['TIPO'].str.replace(
    r'(?<!^)([A-Z])',
    r' - \1',
    regex=True
)


df['NOME'] = df['NOME'].apply(
    lambda x: re.sub(r'(?<!^)([A-Z])', r' \1', x)
)

df.to_csv("dataframes/pokemon_data.csv", index=False)

df
```
